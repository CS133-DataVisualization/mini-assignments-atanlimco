## Load the data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

IGN="https://raw.githubusercontent.com/csbfx/advpy122-data/master/ign_scores.csv"
## Your code here . . .

#read the csv file
ign_scores = pd.read_csv(IGN)

#set platform as index
ign_scores = ign_scores.set_index("Platform")
ign_scores
```

## Problem 5 - (4 points)
Use the Pokemon dataset to create a clustermap with color. First, filter the dataframe to only keep data with `Type 1` equals to following values: `Water`, `Normal`, `Grass`, `Bug` and `Psychic`. Annotate the dendrogram using different colors for these five different `Type 1` values. Use `Name` as the index.

pokemon_data is in https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv

```python
# Problem 5
# Your code here . . .
pokemon_data = "https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv"

#read in the pokemon csv file
pokemon_df = pd.read_csv(pokemon_data)

#filter the dataset to only the requested types
kept_types = pokemon_df[pokemon_df["Type 1"].isin(["Water", "Normal", "Grass", "Bug", "Psychic"])]

#set Name as the index
kept_types = kept_types.set_index("Name")

#drop the stats unrelated to its actual stats
kept_types_stats = kept_types.drop(["Legendary", "Type 2", "Generation", "#"], axis=1)
kept_types_stats

#add the column of colors to annotate the dendrogram based on Type 1
type_color=kept_types_stats["Type 1"].replace({"Water":"blue","Normal":"gray","Grass":"green", "Bug":"lightgreen", "Psychic":"pink"})
kept_types_stats = kept_types_stats.drop(columns=["Type 1"])

#clustermap with the annotation to the dendogram by setting row_colors as type_color
g=sns.clustermap(
    kept_types_stats,
    figsize=(20,20),
    row_colors=type_color
)

#plot title
g.fig.suptitle(
"""Statistics of Water, Normal, Grass, Bug, and Psychic Pokemon
   (blue=Water,gray=Normal,green=Grass,lightgreen="Bug,pink="Psychic)""", y=1.05,
);
```
