Here are the descriptions of this dataset:
- `#` - Pokedex entry number of the Pokemon
- `Name` - name of the Pokemon
- `Type 1` - each Pokemon has a type, this determines weakness/resistance to attacks [referred to as the primary type]
- `Type 2` - some Pokemon are dual type and have 2 [referred to as the secondary type]
- `Total` - sum of all stats that come after this, a general guide to how strong a Pokemon is
- `HP` - hit points, or health, defines how much damage a Pokemon can withstand before fainting
- `Attack` - the base modifier for normal attacks
- `Defense` - the base damage resistance against normal attacks
- `Sp. Atk` - special attack, the base modifier for special attacks
- `Sp. Def` - the base damage resistance against special attacks
- `Speed` - determines which Pokemon attacks first each round
- `Generation` - refers to which grouping/game series the Pokemon was released in
- `Legendary` - a boolean that identifies whether the Pokemon is legendary

## Q1: Plot the relationship between Attack and Defense
Create a relationship plot using seaborn's `replot` function. Plot `Attack` on the x-axis and `Defense` on the y-axis. Use `hue` to display the `HP` (hit point) for each datapoint. You can choose your favorite palette. Include a meaninful title in the plot. After creating the plot, insert a markdown cell immediately following the plot to describe the relationship between Attack and Defense in this dataset.

```python
# Your code here . . .

### Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

csv_Filepath = "https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv"
pkm = pd.read_csv(csv_Filepath)

plt.style.use("dark_background")

sns.relplot(data=pkm, x="Attack", y="Defense", hue="HP", palette="Spectral")
plt.title("Attack vs Defense vs HP Relationship");
```

# From the plot, I observed....

I noticed that generally, the three stats appear to be positively correlated with each other, such that if one stat is higher, then the other two are also likely going to be high. This makes sense, as certain Pokemon are naturally stronger than others and would have better overall stats, meaning Attack, Defense, and HP would be positively correlated.

However, there are a few outliers. For example, there is a Pokemon that has a little over 175 Attack points but only around 25 Defense points. Furthermore, it has roughly 50 to 100 HP. This likely means the Pokemon is a "glass cannon", as it is able to cause a lot of damage, but can be taken out fairly easily.

Another outlier is the one purple point towards the bottom left of the plot. That Pokemon has incredibly low Attack and Defense, but likely the highest HP in the dataset. This likely means the Pokemon is meant to just tank hits rather than deal damage to the opponent and likely has other abilities that aren't related to damage dealing to compensate.

Additionally, there is an outlier in the top left of the plot, whose Attack and HP are incredibly low, but its Defense is one of the highest in the dataset.

## Q2: Plot the relationship between Attack and Total using a hexbin plot
Create a hexbin plot using seaborn's `joinplot` function. Plot `Attack` on the x-axis and `Total` on the y-axis. You can choose the gridsize. Include a title for the plot "Attack vs total for all Pokemon". After generating the plot, insert a markdown cell to answer this question: The hexbin plot displays the relationship between `Attack` and `Total`, what else does this plot show?

```python
# Your code here . . .

plt.style.use("default")

pkm_hex=sns.jointplot(
    data=pkm,
    x="Attack",
    y="Total",
    kind="hex",
    gridsize=20
)

pkm_hex.fig.suptitle("Attack vs total for all Pokemon", y=1.05);
```

# From the plot, I observed....

In addition to the relationship between Attack points and Total points, the hexbin graph also shows that a lot of the Pokemon have around 50 Attack points and 325 Total points. The graph also shows that a lot of Pokemon have between 75-100 Attack points with 475 Total points. Overall, the hexbin plot makes it easier to see the concentration of points on the graph compared to the dot plot.

## Q3: Create two categorical plots for Pokemon types. `Type 1` and `Type 2` .

```python
# Your code here . . .

type_1 = sns.catplot(
    y='Type 1',
    data=pkm,
    kind='count',
    color='red'
).set_axis_labels('# of each Type 1', 'Type 1 Categories');
```

```python
type_2 = sns.catplot(
    y='Type 2',
    data=pkm,
    kind='count',
    color='blue'
).set_axis_labels('# of each Type 2', 'Type 2 Categories');
```
