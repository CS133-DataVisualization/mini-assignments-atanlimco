In this hands-on exercise, you answer the following questions by using pandas data structures and methods to analyze the eukaryote genome data store in the following tab-delimited file. https://raw.githubusercontent.com/csbfx/advpy122-data/master/euk.tsv

```python
import pandas as pd
%matplotlib inline

# this input file is tab-delimited instead of comma-delimited
tsvFile = "https://raw.githubusercontent.com/csbfx/advpy122-data/master/euk.tsv"
# Load the csv file into a dataframe
euk = pd.read_csv(tsvFile, sep="\t")
euk
```

## Q1. How many Mammals have at least 20,000 genes? What are their scientific names?
*Note:*
- *Mammals are under Class*
- *Scientific names are under Species*


```python
# Your code here . . .

#the boolean conditions in the brackets filter for rows where the Class is 'Mammals' and the Number of genes is at least 20000
#the 'Number of genes' column is an object, not an int, so need to do to_numberic
#errors='coerce' is needed to force to_numeric to process dashes
at_least_20000_gene_mammals = euk[(euk.Class == 'Mammals') & (pd.to_numeric(euk['Number of genes'], errors='coerce') >= 20000)]

#Series.unique returns an array of unique values in the series
#so Species.unique returns an array of unique Species name
#and len() returns the length of that array, which would be the number of unique Species names
print("There are {} Mammals with at least 20,000 genes.".format(len(at_least_20000_gene_mammals.Species.unique())))

#This would print out each item in the Species.unique() array
#and each item is a unique Species name
print("Their scientific names are:")
for species in at_least_20000_gene_mammals.Species.unique():
  print("   {}".format(species))

```

### Q2. Animals are a part of Kingdom. How many records are there for each Class of Animals?

```python
# Your code here . . .

#value_counts returns the number of entries for each type of Class whose Kingsom is Animals
euk[euk.Kingdom == 'Animals'].Class.value_counts()
```

### Q3. Animals are a part of Kingdom. How many unique Species are there for each Class of Animals?

```python
# Your code here . . .

#returns an array of unique Classes whose kingdom is Animals
classes = euk[euk.Kingdom == 'Animals'].Class.unique()

#euk.Class == c filters out euk to only include entries whose Class is c
#Series.nunique() returns the number of unique species
for c in classes:
    print(c, euk[euk.Class == c].Species.nunique())
```

### Q4. What are the unique scientific names of Mammals with the genus name Macaca?
*Recall: the scientific name starts with genus followed by a space and then the species name. Example: Homo sapiens, where Homo is the genus name and sapiens is the species name.*

```python
# Your code here . . .

#I think the .str lets you perform string operations on the contents of the Species column,
#and the Species column is a Series object?
macaca_data = euk[(euk.Class == 'Mammals') & (euk.Species.str.split(' ').str[0] == "Macaca")]

#Species.unique returns an array of unique Species
unique_macaca = macaca_data.Species.unique()

#prints out each species in unique_macaca
for macaca in unique_macaca:
  print(macaca)
```

### Q5. Modify the Species names to only contain the scientific names  and create a new dataframe.

Some of the names in the Species column have more than two parts, such as `Emiliania huxleyi CCMP1516`. Create a new column `Species` that contains only the first two parts of name, such as `Emiliania huxleyi`. Combine this new Species column with `Kingdom`, `Class`, `Size (Mb)`, `Number of genes`, and `Number of proteins` and store this new dataframe as `df_species`.

Hint: Follow Q3 in Lecture 4 with a little twist. Instead of just getting the first element from the split results, you will get the first two elements using `.str[0:2]` which will give you a list. You can then use `.str.join(" ")` to change it back to a string.

```python
# Your code here . . .

#.str lets you perform string operations on a Series (like a column)
#we're setting the Species column of euk to a modified version of Species
#where each Species entries has been turned into an array based on whitespace (split(" "))
#and only the first two items in the array are kept (str[0:2])
#and joined together again with whitespace (str.join(" "))
euk['Species']=euk.Species.str.split(" ").str[0:2].str.join(" ")

#puts the Species, Kingdom, Class, Size (Mb), Number of genes, and Number of proteins columns into df_species
df_species = euk[["Species", "Kingdom", "Class", "Size (Mb)", "Number of genes", "Number of proteins"]]
df_species
```

### Q6. Create a pie plot using pandas to show the number of unique Species in each Class of Animals using the new dataframe you created in Q5
Hint: First, create a new dataframe that contains the number of unique Species and the index is the corresponding Animals Class. Then, use that dataframe to plot the pie plot.
[Check out this documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.pie.html)

```python
# Create a new dataframe that contains the number of unique Species and
# the index is the corresponding Animals Class.

# Your code here . . .
unique_species_dict = {}

#Class.unique() returns an array of unique class names
classes = df_species[df_species.Kingdom == 'Animals'].Class.unique()

#populating a dictionary with the class name as the key and the number of [unique species with that class] as the value
for c in classes:
  unique_species_dict.update({c : df_species[df_species.Class == c].Species.nunique()})

#create a DataFrame where the dictionary values are the data and the dictionary keys are the index
#labeled the data column as "Number of Unique Species" and index as "Class"
pie_plot_df = pd.DataFrame(data=unique_species_dict.values(), index=unique_species_dict.keys(), columns=['Number of Unique Species'])
pie_plot_df.index.names = ['Animals Class']
pie_plot_df
```

```python
# Create a panda pie plot using the dataframe above

# Your code here . . .

pie_plot_df.plot.pie(y='Number of Unique Species')
```
